#### Valores faltantes

Antes de avanzar, es importante verificar si hay datos incompletos en el dataset.  
Usamos `isnull().sum()` para contar los valores faltantes por columna.  


```python
# Conteo de valores faltantes
titanic_df.isnull().sum()
```

#### Tratamiento de valores faltantes

En el dataset del Titanic detectamos tres columnas con datos incompletos:

- **Edad (`Age`)**: faltan 177 valores (~20%).  
  Se completarán usando la **mediana de las edades** (28 años), ya que este método no se ve afectado por valores extremos como bebés o ancianos.  

- **Cabina (`Cabin`)**: faltan 687 valores (~77%).  
  Debido a que la mayoría de los registros está vacío, se **eliminará esta columna** en el análisis exploratorio.  

- **Puerto de embarque (`Embarked`)**: faltan 2 valores (<1%).  
  Se completarán con la **moda** (el valor más frecuente), que corresponde a **Southampton (`S`)**.  

Después de aplicar estos pasos, el dataset queda **sin valores faltantes** y listo para continuar con los análisis y visualizaciones.


```python
titanic_clean = titanic_df.copy()

titanic_clean["Age"].fillna(titanic_clean["Age"].median(), inplace=True)

titanic_clean.drop(columns=["Cabin"], inplace=True)

titanic_clean["Embarked"].fillna(titanic_clean["Embarked"].mode()[0], inplace=True)

titanic_clean.isnull().sum()
```

```python
titanic_clean.describe().T
```

#### Distribución de supervivientes

La variable `Survived` indica quién sobrevivió (1) y quién no (0).  
El promedio de esta variable fue **0.38**, lo que confirma que **solo el 38% de los pasajeros sobrevivió**.  

Podemos visualizarlo con un gráfico de barras:


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x="Survived", data=titanic_clean, palette="Set2")
plt.title("Distribución de supervivientes")
plt.xlabel("Supervivencia (0 = No, 1 = Sí)")
plt.ylabel("Cantidad")
plt.show()
```

El gráfico muestra que la mayoría de pasajeros **no sobrevivió** al naufragio.  
En total, aproximadamente **550 personas fallecieron** frente a unas **340 que sobrevivieron**.  

Esto concuerda con las estadísticas iniciales: solo el **38% de los pasajeros sobrevivió**.  
Por lo tanto, la variable `Survived` está **desbalanceada**, con más casos de fallecidos que de sobrevivientes.


#### Distribución por clase de pasajero

La variable `Pclass` representa la clase de ticket (1ª, 2ª, 3ª).  
La media fue **2.3**, lo que significa que la mayoría de pasajeros viajaba en **segunda o tercera clase**.  

Visualizamos la cantidad de pasajeros por clase:


```python
sns.countplot(x="Pclass", data=titanic_clean, palette="Set3")
plt.title("Distribución de pasajeros por clase")
plt.xlabel("Clase de pasajero")
plt.ylabel("Cantidad")
plt.show()
```

El gráfico indica que:  
- La **mayoría de pasajeros viajaba en tercera clase** (cerca de 500).  
- Menos personas viajaban en primera (~210) y segunda clase (~180).  

Esto coincide con la media de `Pclass` = 2.3, que sugiere que la mayor parte de los pasajeros pertenecía a **clases sociales más bajas**.  

Este resultado es importante, ya que nos permitirá comprobar después si la clase influyó en las probabilidades de supervivencia.


#### Hipótesis 1: Las mujeres tuvieron más probabilidades de sobrevivir

Para verificar esta hipótesis, comparamos la variable `Sex` con `Survived`.  
El gráfico muestra cuántos hombres y mujeres sobrevivieron o no.


```python
sns.countplot(x="Sex", hue="Survived", data=titanic_clean, palette="Set1")
plt.title("Supervivencia por género")
plt.xlabel("Género")
plt.ylabel("Cantidad")
plt.legend(title="Supervivió", labels=["No", "Sí"])
plt.show()

```

El gráfico confirma que existió una diferencia marcada en la supervivencia según el género:

- En el caso de los **hombres**, la gran mayoría **no sobrevivió**. Solo un grupo reducido logró salvarse.  
- En el caso de las **mujeres**, la mayoría **sí sobrevivió**, mientras que un grupo menor falleció.  

Esto valida la hipótesis inicial: **el género influyó de forma decisiva en la supervivencia**.


### Hipótesis 2: Si viajabas en primera clase tenías más probabilidades de sobrevivir

Ahora analizaremos si la clase social influyó en la probabilidad de supervivencia. 
Para esta hipótesis analizamos la variable `Pclass` frente a `Survived`.  
El gráfico compara la supervivencia en 1ª, 2ª y 3ª clase.  


```python
sns.countplot(x="Pclass", hue="Survived", data=titanic_clean, palette="Set3")
plt.title("Supervivencia por clase de pasajero")
plt.xlabel("Clase de pasajero")
plt.ylabel("Cantidad")
plt.legend(title="Supervivió", labels=["No", "Sí"])
plt.show()

```

Este resultado confirma la hipótesis inicial: **los pasajeros de primera clase tuvieron más probabilidades de sobrevivir que los de tercera clase**, lo cual refleja el fuerte impacto de las condiciones socioeconómicas en este desastre.


### Distribución de edades

La variable `Age` nos permite explorar el rango de edades de los pasajeros.  
Además, podemos analizar si la edad influyó en la supervivencia.

- Entre los **adultos jóvenes (20–40 años)** predominan los fallecidos frente a los sobrevivientes.  
- En las edades más avanzadas (50 años en adelante), la supervivencia fue mucho menor en comparación con la mortalidad.  

En conclusión, la **edad influyó en las probabilidades de supervivencia**: los más jóvenes, especialmente los niños, tuvieron mejores oportunidades de sobrevivir en comparación con los adultos mayores.


### Matriz de correlación

Analizamos las relaciones entre variables numéricas (`Age`, `SibSp`, `Parch`, `Fare`).  
Una correlación cercana a **1** significa relación positiva fuerte; cercana a **-1**, relación negativa fuerte.  


```python
# Calcular y visualizar matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(titanic_clean.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación - Titanic')
plt.show()
```

Analizamos las relaciones entre las variables numéricas del dataset (`Age`, `SibSp`, `Parch`, `Fare`) y su vínculo con la supervivencia (`Survived`).  
La correlación mide la fuerza y dirección de la relación entre dos variables:  
- Valores cercanos a **1** → correlación positiva fuerte.  
- Valores cercanos a **-1** → correlación negativa fuerte.  
- Valores cercanos a **0** → sin relación clara.

La gráfica indica:

- **Clase del pasajero (`Pclass`)**: presenta una correlación negativa con la supervivencia (`-0.34`).  
  Esto significa que a mayor clase numérica (es decir, peor clase: 3ª), menor probabilidad de sobrevivir. En cambio, los pasajeros de **1ª clase** tuvieron más posibilidades de sobrevivir.

- **Tarifa pagada (`Fare`)**: correlación positiva con `Survived` (`0.26`).  
  Los pasajeros que pagaron tarifas más altas (normalmente de primera clase) presentaron mayor supervivencia. Este hallazgo refuerza la relación entre **estatus económico** y **probabilidades de sobrevivir**.

- **Número de hermanos/esposos (`SibSp`) y padres/hijos (`Parch`)**: muestran una correlación moderada entre sí (`0.41`), lo cual refleja que algunos pasajeros viajaban en familias.  
  Sin embargo, la relación de estas variables con la supervivencia es débil, lo que sugiere que **viajar acompañado no garantizaba sobrevivir**.

- **Edad (`Age`)**: no muestra una correlación fuerte con la supervivencia (valor cercano a `0`).  
  Aunque sí presenta correlaciones negativas con `SibSp` y `Parch`, indicando que a mayor edad, los pasajeros tenían menos familiares a bordo.


Las variables que más influyeron en la supervivencia fueron la **clase del pasajero (`Pclass`)** y la **tarifa pagada (`Fare`)**, ambas relacionadas con el nivel socioeconómico del viajero. Factores como la edad o viajar en familia tuvieron un efecto mucho menor en comparación.

La **clase social** (reflejada en `Pclass` y `Fare`) fue el factor más determinante en la supervivencia. Aunque la edad y la presencia de familiares tuvieron algún efecto, su impacto fue mucho menor comparado con la **posición socioeconómica del pasajero**.

Esta gráfica refuerza que la **clase social** y la **tarifa pagada** fueron factores críticos.

En general, el **factor económico (tarifa / clase)** tuvo más peso que la edad en las probabilidades de supervivencia.
