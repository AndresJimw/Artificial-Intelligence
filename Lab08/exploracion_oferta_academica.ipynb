```python
df = pd.read_excel(xlsx_path, sheet_name=0, header=13)

# Limpiamos filas completamente vacías
df = df.dropna(how="all")

# limpiamos espacios en los nombres de columna
df.columns = [str(c).strip() for c in df.columns]

df.head()
```

```python
df.shape, df.columns
```

```python
# 1) Normalizamos texto de columnas relevantes
df["ESTADO"] = df["ESTADO"].astype(str).str.strip()
df["MODALIDAD"] = df["MODALIDAD"].astype(str).str.upper().str.strip()
df["NIVEL_FORMACIÓN"] = df["NIVEL_FORMACIÓN"].astype(str).str.upper().str.strip()

# 2) Filtro de instituciones/carreras VIGENTES
mask_vigente = df["ESTADO"].str.startswith("VIGENTE")

# 3) Filtro de modalidad EN LÍNEA (con o sin tilde)
mask_en_linea = df["MODALIDAD"].str.contains("EN LINEA") | df["MODALIDAD"].str.contains("EN LÍNEA")

# 4) Filtro de CUARTO NIVEL / POSGRADO
mask_posgrado = df["NIVEL_FORMACIÓN"].str.contains("CUARTO NIVEL") | \
                 df["NIVEL_FORMACIÓN"].str.contains("POSGRADO") | \
                 df["NIVEL_FORMACIÓN"].str.contains("EDUCACIÓN SUPERIOR DE POSGRADO")

# 5) Aplicamos los tres filtros
df_online_vig_pos = df[mask_vigente & mask_en_linea & mask_posgrado].copy()

df_online_vig_pos.head()
```

```python
# 1) Normalizamos texto de columnas relevantes
df["ESTADO"] = df["ESTADO"].astype(str).str.strip()
df["MODALIDAD"] = df["MODALIDAD"].astype(str).str.upper().str.strip()
df["NIVEL_FORMACIÓN"] = df["NIVEL_FORMACIÓN"].astype(str).str.upper().str.strip()
df["PROVINCIA"] = df["PROVINCIA"].astype(str).str.upper().str.strip()   # <-- NUEVO

# 2) Filtro de instituciones/carreras VIGENTES
mask_vigente = df["ESTADO"].str.startswith("VIGENTE")

# 3) Filtro de modalidad EN LÍNEA (con o sin tilde)
mask_en_linea = df["MODALIDAD"].str.contains("EN LINEA") | df["MODALIDAD"].str.contains("EN LÍNEA")

# 4) Filtro de CUARTO NIVEL / POSGRADO
mask_posgrado = (
    df["NIVEL_FORMACIÓN"].str.contains("CUARTO NIVEL") |
    df["NIVEL_FORMACIÓN"].str.contains("POSGRADO") |
    df["NIVEL_FORMACIÓN"].str.contains("EDUCACIÓN SUPERIOR DE POSGRADO")
)

# 5) Filtro de provincias (solo Pichincha o Guayas)
mask_prov = df["PROVINCIA"].isin(["PICHINCHA"])

# 6) Aplicamos todos los filtros
df_online_vig_pos = df[mask_vigente & mask_en_linea & mask_posgrado & mask_prov].copy()

df_online_vig_pos.head()
```

```python
# Guardar el CSV filtrado dentro de la carpeta data
output_path = "./data/posgrados_pichincha_en_linea.csv"

df_online_vig_pos.to_csv(output_path, index=False, encoding="utf-8")

output_path
```
